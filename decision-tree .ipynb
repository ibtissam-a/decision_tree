```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../input/titanic-dataset/Titanic-Dataset.csv')
df.head(2)
```

```python
target =df[['Survived']]
target.head(3)
```

```python
# we will only work with 'Pclass','Sex','Age','Fare' columns in order to predict the survived persons 
df2 = df[['Pclass','Sex','Age','Fare']]
df2.head(2)
```

```python
# we will deal with the "Sex" column values
df3 = pd.get_dummies(df2.Sex)
dataframe = pd.concat([df2, df3],axis='columns')
dataframe.drop('Sex', axis='columns', inplace=True)
dataframe.head(3)
```

```python
dataframe.info() 
# we can see that there is missed values in column "Age"
```

```python
# we used the "ffill" to mange the missed values 
new_df = dataframe.fillna(method="ffill")
new_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_df,target,test_size=0.2)
```

```python
# create our tree
model_tree = DecisionTreeClassifier()
```

```python
model_tree.fit(x_train, y_train )
```

```python
# predict the model score 
model_tree.score(x_test, y_test )
```
